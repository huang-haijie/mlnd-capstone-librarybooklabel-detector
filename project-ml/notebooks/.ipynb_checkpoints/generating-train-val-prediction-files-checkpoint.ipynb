# Generating Train-Val Prediction Files

This notebook shows the scripts to generate prediction outputs (Darkflow JSON format) on training and validation sets by loading the weights from specified checkpoints.  The checkpoints were saved during the training process.

**Inputs**:
- Model config file
- Folder containing Tensorflow training checkpoints
- Folders containing images of training and validation sets

**Outputs**:
- Darkflow prediction output files arranged in the below folder structure:
    - [epoch_number_1]
        - train-out
            - [image_file_1].json
            - [image_file_2].json
            - ...
        - val-out
            - [image_file_1].json
            - [image_file_2].json
            - ...
    - [epoch_number_2]
    - ...

The generated prediction output files will be used to compare against the ground truth annotations to calculate the AP scores and evaluation results for each checkpoint.

```python
from darkflow.net.build import TFNet
import os
import shutil
import glob
```

## Training Checkpoint Predictions on COCO-Text

```python
## Functions to generate prediction outputs using specified checkpoints

def move_files(src_dir, dst_dir):
    srcs = glob.glob(os.path.join(src_dir, '*'))
    for src in srcs:
        shutil.move(src, dst_dir)

def generate_predictions_by_ckptlist(ckptlist):
    for step in ckptlist:
        # load checkpoint
        load = step
        tfnet.FLAGS.load = load
        tfnet.load_from_ckpt()
        n_epoch = int(load * batch / n_train_img)
        
        # generate predictions for training set
        imgdir = imgdir_train
        tfnet.FLAGS.imgdir = imgdir
        tfnet.predict()
        output_dir = os.path.join(imgdir, 'out')
        dst_dir = os.path.join(train_predictions_dir, str(n_epoch)+'/train-out')
        os.makedirs(dst_dir, exist_ok=True)
        move_files(output_dir, dst_dir)
        
        # generate predictions for validation set
        imgdir = imgdir_val
        tfnet.FLAGS.imgdir = imgdir
        tfnet.predict()
        output_dir = os.path.join(imgdir, 'out')
        dst_dir = os.path.join(train_predictions_dir, str(n_epoch)+'/val-out')
        os.makedirs(dst_dir, exist_ok=True)
        move_files(output_dir, dst_dir)
        
        print('Done.')
        
# Generate predictions on training and validation sets using the specified checkpoints.
def generate_predictions(start, stop, step):
    for step in range(start, stop+1, step):
        # load checkpoint
        load = step
        tfnet.FLAGS.load = load
        tfnet.load_from_ckpt()
        n_epoch = int(load * batch / n_train_img)
        
        # generate predictions for training set
        imgdir = imgdir_train
        tfnet.FLAGS.imgdir = imgdir
        tfnet.predict()
        output_dir = os.path.join(imgdir, 'out')
        dst_dir = os.path.join(train_predictions_dir, str(n_epoch)+'/train-out')
        os.makedirs(dst_dir, exist_ok=True)
        move_files(output_dir, dst_dir)
        
        # generate predictions for validation set
        imgdir = imgdir_val
        tfnet.FLAGS.imgdir = imgdir
        tfnet.predict()
        output_dir = os.path.join(imgdir, 'out')
        dst_dir = os.path.join(train_predictions_dir, str(n_epoch)+'/val-out')
        os.makedirs(dst_dir, exist_ok=True)
        move_files(output_dir, dst_dir)
        
        print('Done.')
        
```

```python
model = '../checkpoints/tiny-yolo-cocotext/yolov2-tiny-voc-1c.cfg'
backup = '../checkpoints/tiny-yolo-cocotext/ckpt'
imgdir_train = '/home/haijie/machine-learning/temp/usb-ml/datasets/COCOText/2017/images/train'
imgdir_val = '/home/haijie/machine-learning/temp/usb-ml/datasets/COCOText/2017/images/val'
imgdir = imgdir_train
labels = '/home/haijie/machine-learning/temp/usb-ml/datasets/COCOText/labels.txt'
load = -1
batch = 32
n_train_img = 43686
train_predictions_dir = '../checkpoints/tiny-yolo-cocotext/train-predictions'
```

```python
# Initialize Darkflow
options = {'model': model, 'load': load, 'backup': backup, 'imgdir': imgdir, 
           'labels': labels, 'batch': batch, 'threshold': 0.1, 'json': True, 'gpu': 0.8}

tfnet = TFNet(options)
```

```python
ckptlist = [13650, 40950, 68250, 95550, 109200, 136500]
generate_predictions_by_ckptlist(ckptlist)
```
