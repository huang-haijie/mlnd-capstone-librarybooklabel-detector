# Data Exploration on Self-collected Library Books Images

## Validation on Annotations Generated by VoTT

The library books images were self-collected from library using mobile phone camera.  The annotations of bounding boxes for book labels are manually tagged using VoTT.  The output annotation files are in Pascal VOC xml format.

```python
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
from lxml import etree as et
from matplotlib.collections import PatchCollection
from matplotlib.patches import Rectangle, PathPatch
import glob
from ipywidgets import widgets, interact, IntSlider
%matplotlib inline
```

```python
def visualize_voc_annotation(img_file_name, img_dir, ann_dir):
    img_file_path = os.path.join(img_dir, img_file_name)
    ann_file_path = os.path.join(ann_dir, img_file_name.split('.')[0] + '.xml')

    # Load and display image
    img = cv2.imread(img_file_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10,8))
    plt.imshow(img)
    show_voc_bbox(ann_file_path)
    
def show_voc_bbox(ann_file_path):
    tree = et.parse(ann_file_path)
    boxes = []
    colors = []
    for obj in tree.iter('object'):
        bndbox = obj.find('bndbox')
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)
        width = xmax - xmin
        height = ymax - ymin
        print(xmin, ymin, width, height)
        boxes.append(Rectangle([xmin,ymin],width,height))
        #colors.append(np.random.random((1,3))[0])
    
    collection = PatchCollection(boxes, facecolors='none', edgecolors='yellow', linewidths=3)
    ax = plt.gca()
    ax.add_collection(collection)
        
```

```python
img_dir = '../../data/librarybooks/images/train'
ann_dir = '../../data/librarybooks/annotations/train'

img_paths = glob.glob(os.path.join(img_dir, '*.jpg'))
print(len(img_paths))
test_img_path = np.random.choice(img_paths)
test_img_name = test_img_path.split('/')[-1]
print(test_img_name)
visualize_voc_annotation(test_img_name, img_dir, ann_dir)
```

```python
# Visualize annotation interactively
def visualize_annotation_by_index(i):
    test_img_path = img_paths[i]
    test_img_name = test_img_path.split('/')[-1]
    print(test_img_name)
    visualize_voc_annotation(test_img_name, img_dir, ann_dir)

interact(visualize_annotation_by_index, i=IntSlider(value=0, description='Index', max=len(img_paths)-1, min=0))
```
